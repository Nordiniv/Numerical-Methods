```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math

x = np.array([1, 1.5, 2, 2.5, 3, 4])
y = np.array([[0.27, 0.64, 1.02, 1.39, 1.77, 2.5]]).T

def neq(x,y):
    x = np.array([np.ones_like(x),x]).T
    c = np.linalg.inv(x.T@x)@(x.T@y)
    normie = np.linalg.norm(c.T@x.T - y.T)
    return c, normie

c, normie = neq(x, y)
display(Math("y(x) = {:.6f}x {:+.6f} , \| d \| = {}".format(*c[1], *c[0], normie)))

def plot(b=0, e=5, s=0.01, f = lambda r: c[0] + r*c[1], re = False):
    ar = np.arange(b, e, s)
    plt.plot(ar, f(ar) );
    plt.plot(x, y, 'o');
    if re == True:
        return f 
    
plot(0, 5, f = lambda r: c[0] + r*c[1])
```

$y = ax^2 + bx + c  \Rightarrow  \mathbf{y} = a \mathbf{x^2} + b \mathbf{x} + c \Rightarrow \begin{bmatrix} 3.14 & 4.15 & 5.65 \end{bmatrix}  = a \begin{bmatrix} 1.4^2 \\ 1.3^2 \\ 1.1^2 \end{bmatrix} + b \begin{bmatrix} 1.4 \\ 1.3 \\ 1.1 \end{bmatrix} +c \begin{bmatrix} 1 \\ 1 \\ 1 \end{bmatrix}= \begin{bmatrix} a & b & c \end{bmatrix} \begin{bmatrix} 1.4^2 & 1.3^2 & 1.1^2 \\ 1.4 & 1.3 & 1.1 \\ 1 & 1 & 1 \end{bmatrix} \\ \Rightarrow \begin{bmatrix} a & b & c \end{bmatrix} = \begin{bmatrix} 3.14 & 4.15 & 5.65 \end{bmatrix}  \begin{bmatrix} 1.4^2 & 1.3^2 & 1.1^2 \\ 1.4 & 1.3 & 1.1 \\ 1 & 1 & 1 \end{bmatrix} ^{-1} $

```python
def interpolate(xs, ys):
    x = np.array([xs])
    y = np.array([ys])
    c = y@np.linalg.inv(np.vstack(tuple(x**i for i in range(len(xs)-1, -1, -1))))
    return c

x = [1.4, 1.3, 1.1]
y = [3.14, 4.15, 5.65]
c = interpolate(x,y)[0]
display(Math(f'P(x) = {float(c[0]):.3f}x^2 + {float(c[1]):.3f}x + {float(c[2]):.3f}'))
plot(0.5, 2, f = lambda r: c[0]*r**2+c[1]*r+c[2])
```

```python
x = [1, 4, 9, 16, 25]
y = np.array([0, 2.89, 5.75, 8.65, 11.53])
def divided_diff(x, y):
    '''
    function to calculate the divided
    differences table
    '''
    x = np.array(x)
    y = np.array(y)
    n = len(y)
    coef = np.zeros([n, n])
    # the first column is y
    coef[:,0] = y
    
    for j in range(1,n):
        for i in range(n-j):
            coef[i][j] = \
           (coef[i+1][j-1] - coef[i][j-1]) / (x[i+j]-x[i])
            
    return coef

def newton_poly(coef, x_data, x):
    '''
    evaluate the newton polynomial 
    at x
    '''
    n = len(x_data) - 1 
    p = coef[n]
    for k in range(1,n+1):
        p = coef[n-k] + (x -x_data[n-k])*p
    return p

def gep(x, y, s, e, x_):
    x_ = np.array(x_)
    # get the divided difference coef
    a_s = divided_diff(x, y)[0, :]

    # evaluate on new data points
    x_new = np.arange(s, e, .1)
    y_new = newton_poly(a_s, x, x_new)

    plt.figure(figsize = (12, 8))
    plt.plot(x_new, y_new);
    plt.plot(x, y, 'o');
    y_ = newton_poly(a_s, x, x_)
    plt.plot(x_, y_, 'rx')
    for i, j in zip(x_, y_):
        display(Math(f'y({i:.1f}) = {j}'))
    return y_
        

y_ = gep(x, y, 0, 26, [3, 10])
```

# $Excersise \ 7$ #

```python
from numpy import sqrt, sin, exp
def rk4(f, t0, y0, h, a):
    from scipy.special import gamma
    ha = (h**a)/gamma(1+a)
    k1 = h*f(t0, y0)
    k2 = h*f(t0 + ha/2, y0 + k1/2)
    k3 = h*f(t0 + ha/2, y0 + k2/2)
    k4 = h*f(t0 + ha, y0 + k3)
    return y0 + (k1 + 2*k2 + 2*k3 + k4)/6

def iterk4(f = lambda x, y : x**2 + y**2, t0 = 0, tf = 1, y0 = 0, h = 0.2, a = 1): 
    for i in range(int((tf-t0)/h)):
        y0 = rk4(f, t0, y0, h, a)
        t0 += h
        display(Math(f'y({t0:.1f}) = {y0}'))

iterk4(lambda x, y: sqrt(10*x**2 + y), 0.2, 0.4, 0.41)
```

```python
iterk4(lambda x, y: sqrt(y/x) + (y/x), 1, 1.1, 1, 0.1)
```

```python
iterk4(lambda x, y: x*y + sqrt(y), 0, 0.4, 1, 0.2)
iterk4(lambda x, y: x*y + sqrt(y), 0, -0.2, 1, -0.2)
```
