```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math

x = np.array([1, 1.5, 2, 2.5, 3, 4])
y = np.array([[0.27, 0.64, 1.02, 1.39, 1.77, 2.5]]).T

def neq(x,y):
    x = np.array([np.ones_like(x),x]).T
    c = np.linalg.inv(x.T@x)@(x.T@y)
    normie = np.linalg.norm(c.T@x.T - y.T)
    return c, normie

c, normie = neq(x, y)
display(Math("y(x) = {:.6f}x {:+.6f} , \| d \| = {}".format(*c[1], *c[0], normie)))

def plot(b=0, e=5, s=0.01, f = lambda r: c[0] + r*c[1], re = False):
    ar = np.arange(b, e, s)
    plt.plot(ar, f(ar) );
    plt.plot(x, y, 'o');
    if re == True:
        return f 
    
plot(0, 5, f = lambda r: c[0] + r*c[1])
```

$ln(y) = ln(bx^a) = ln(b) + a \cdot ln(x) \\Y = aX + B$

```python
x = np.array([2, 3, 5, 6, 10])
y = np.array([[2.83, 5.2, 11.18, 14.7, 31.62]]).T
X = np.log(x)
Y = np.log(y)
c, normie = neq(X, Y)
display(Math("Y(X) = {:.6f}X {:+.6f} , \| d \| = {:.6f}".format(*c[1], *c[0], normie)))
display(Math("y(x) = " + f"{np.exp(*c[0]):.6f}" + "x^{" + f"{float(*c[1]):.6f}" + "}"))
plot(0, 11, f = lambda r: np.exp(c[0])*(r)**(c[1]))
```

$\sqrt{y} =\bigg| \frac{1}{b} + \frac{a}{b} x^{-\frac{1}{2}} \bigg| \rightarrow \sqrt{y} =\bigg| c + d x^{-\frac{1}{2}} \bigg| \ : c = \frac{1}{b}, d = \frac{a}{b}$

$ since \ y(x) \ is \ defined \ using \sqrt x \, then \   D= (0, \inf) \ and \ \sqrt{y} =c + d x^{-\frac{1}{2}} = Y = c + d X$

```python
x = np.array([2.5, 3, 3.5, 4, 4.5])
y = np.array([[7.45, 6.3, 5.47, 4.84, 4.35]]).T
Y = np.sqrt(y)
X = x**-.5
theta, normie = neq(X, Y)
c, d = *theta[0], *theta[1]
b = 1/c
a = b*d
display(Math(r'y(x)= \bigg( \frac{ \sqrt{x} + ' + f'{a:.6f}' + '}{' + f'{b:.6f}' + r'\sqrt{x}} \bigg)^2 \ , \|d\| =' + f'{normie:.6f}' ))
plot(0.5, 6, f = lambda r: ((np.sqrt(r)+a)/(b*np.sqrt(r)))**2)
```

$y = ax^2 + bx + c  \Rightarrow  \mathbf{y} = a \mathbf{x^2} + b \mathbf{x} + c \Rightarrow \begin{bmatrix} 3.14 & 4.15 & 5.65 \end{bmatrix}  = a \begin{bmatrix} 1.4^2 \\ 1.3^2 \\ 1.1^2 \end{bmatrix} + b \begin{bmatrix} 1.4 \\ 1.3 \\ 1.1 \end{bmatrix} +c \begin{bmatrix} 1 \\ 1 \\ 1 \end{bmatrix}= \begin{bmatrix} a & b & c \end{bmatrix} \begin{bmatrix} 1.4^2 & 1.3^2 & 1.1^2 \\ 1.4 & 1.3 & 1.1 \\ 1 & 1 & 1 \end{bmatrix} \\ \Rightarrow \begin{bmatrix} a & b & c \end{bmatrix} = \begin{bmatrix} 3.14 & 4.15 & 5.65 \end{bmatrix}  \begin{bmatrix} 1.4^2 & 1.3^2 & 1.1^2 \\ 1.4 & 1.3 & 1.1 \\ 1 & 1 & 1 \end{bmatrix} ^{-1} $

```python
def interpolate(xs, ys):
    x = np.array([xs])
    y = np.array([ys])
    c = y@np.linalg.inv(np.vstack(tuple(x**i for i in range(len(xs)-1, -1, -1))))
    return c

x = [1.4, 1.3, 1.1]
y = [3.14, 4.15, 5.65]
c = interpolate(x,y)[0]
display(Math(f'P(x) = {float(c[0]):.3f}x^2 + {float(c[1]):.3f}x + {float(c[2]):.3f}'))
plot(0.5, 2, f = lambda r: c[0]*r**2+c[1]*r+c[2])
```
